## Business Problem
An e-commerce company wants to understand its customer revenue distribution, identify loyal vs. one-time buyers, and uncover which product categories and time periods drive the most revenue.

## Business Questions
1. Do 20% of customers drive 80% of revenue? *(Pareto Analysis)*
2. What is the repeat purchase rate — and how do loyal buyers compare to one-time buyers?
3. Which months and quarters drive the most revenue? *(Seasonality)*
4. Which product categories perform best by revenue and volume?
5. Who are the Top 10 customers by lifetime value? *(SQL Window Functions)*

```
# ── SECTION 1: IMPORTS & SETUP ──────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import sqlite3
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (13, 6)
plt.rcParams['font.family']    = 'sans-serif'
plt.rcParams['axes.spines.top']   = False
plt.rcParams['axes.spines.right'] = False
sns.set_palette("Blues_d")

print("✓ Libraries loaded.")
```

```
# ── SECTION 2: GENERATE SYNTHETIC RETAIL DATASET ────────────
np.random.seed(42)

# --- Customers ---
N_CUST = 500
cust_ids = [f'CUST_{i:04d}' for i in range(1, N_CUST + 1)]
cities   = ['New York','Los Angeles','Chicago','Houston','Phoenix',
            'Philadelphia','San Antonio','San Diego','Dallas','San Jose',
            'Austin','Columbus','Charlotte','Indianapolis','Seattle']

customers = pd.DataFrame({
    'customer_id':   cust_ids,
    'customer_city': np.random.choice(cities, N_CUST),
    'signup_date':   pd.to_datetime(
        np.random.choice(pd.date_range('2022-01-01','2023-06-30'), N_CUST))
})

# --- Products (50 SKUs across 5 categories) ---
cat_price = {'Electronics':(100,900),'Clothing':(20,150),
             'Home & Garden':(15,300),'Sports':(25,250),'Books':(8,60)}
products = []
for i in range(1, 51):
    cat = np.random.choice(list(cat_price))
    lo, hi = cat_price[cat]
    products.append({'product_id':f'PROD_{i:03d}',
                     'product_name':f'{cat.split()[0]}_Item_{i}',
                     'category':cat,
                     'unit_price':round(np.random.uniform(lo, hi), 2)})
products_df = pd.DataFrame(products)

# --- Orders (Pareto-skewed: top 20% customers ~70% of orders) ---
weights = np.random.pareto(1.8, N_CUST) + 0.1
weights /= weights.sum()

def seasonal_date():
    mw = [0.05,0.05,0.07,0.07,0.08,0.08,0.07,0.08,0.08,0.09,0.10,0.18]
    m  = np.random.choice(range(1,13), p=mw)
    y  = np.random.choice([2023, 2024])
    d  = np.random.randint(1, 28)
    return f'{y}-{m:02d}-{d:02d}'

N_ORD = 3000
orders = pd.DataFrame({
    'order_id':    [f'ORD_{i:05d}' for i in range(1, N_ORD+1)],
    'customer_id': np.random.choice(cust_ids, N_ORD, p=weights),
    'order_date':  pd.to_datetime([seasonal_date() for _ in range(N_ORD)])
}).sort_values('order_date').reset_index(drop=True)

# --- Order Items ---
rows = []
for _, o in orders.iterrows():
    for _, p in products_df.sample(np.random.randint(1,5)).iterrows():
        qty = np.random.randint(1, 4)
        rows.append({'order_id':o['order_id'], 'product_id':p['product_id'],
                     'category':p['category'], 'quantity':qty,
                     'unit_price':p['unit_price'],
                     'line_total':round(qty*p['unit_price'],2)})
items_df = pd.DataFrame(rows)

totals = items_df.groupby('order_id')['line_total'].sum().reset_index()
totals.columns = ['order_id','order_total']
orders = orders.merge(totals, on='order_id')

print("Dataset created successfully!")
print(f"  Customers   : {len(customers):,}")
print(f"  Products    : {len(products_df):,}")
print(f"  Orders      : {len(orders):,}")
print(f"  Order Lines : {len(items_df):,}")
print(f"  Total Revenue: ${orders['order_total'].sum():,.2f}")
```

```
# ── SECTION 3: LOAD INTO SQLite ──────────────────────────────
conn = sqlite3.connect(':memory:')
customers.to_sql('customers',  conn, index=False, if_exists='replace')
products_df.to_sql('products', conn, index=False, if_exists='replace')
orders.to_sql('orders',        conn, index=False, if_exists='replace')
items_df.to_sql('order_items', conn, index=False, if_exists='replace')

print("Tables loaded into SQLite in-memory database:")
for t in ['customers','products','orders','order_items']:
    n = pd.read_sql(f"SELECT COUNT(*) AS c FROM {t}", conn)['c'][0]
    print(f"  {t:<15} {n:>6,} rows")
```

## Analysis 1 · Pareto Revenue Analysis
**Business Question:** Do 20% of customers drive the majority of revenue?

Uses `NTILE(5)` SQL window function to bucket customers into quintiles.

```
pareto_sql = '''
WITH cust_rev AS (
    SELECT customer_id,
           COUNT(DISTINCT order_id)                            AS orders,
           ROUND(SUM(order_total), 2)                          AS revenue
    FROM   orders
    GROUP  BY customer_id
),
bucketed AS (
    SELECT customer_id, orders, revenue,
           NTILE(5) OVER (ORDER BY revenue DESC)               AS quintile
    FROM   cust_rev
)
SELECT quintile,
       COUNT(*)                                                AS customers,
       ROUND(SUM(revenue), 2)                                  AS group_revenue,
       ROUND(SUM(revenue)*100.0 / SUM(SUM(revenue)) OVER (), 1) AS pct_total
FROM   bucketed
GROUP  BY quintile
ORDER  BY quintile
'''
pareto = pd.read_sql(pareto_sql, conn)
print(pareto.to_string(index=False))
top20 = pareto[pareto['quintile']==1]['pct_total'].values[0]
print(f"\n>>> Key Finding: Top 20% of customers = {top20}% of total revenue.")
```

```
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
blues = ['#1a3a5c','#2e6da4','#4e9fd4','#89c4e1','#c6e2f0']

bars = axes[0].bar([f'Q{q}\n(Top {q*20}%)' for q in pareto['quintile']],
                   pareto['pct_total'], color=blues)
axes[0].set_title('Revenue Share by Customer Quintile', fontsize=14, fontweight='bold', pad=15)
axes[0].set_ylabel('% of Total Revenue')
for b, v in zip(bars, pareto['pct_total']):
    axes[0].text(b.get_x()+b.get_width()/2, b.get_height()+.5, f'{v}%',
                 ha='center', fontweight='bold')

cum = [0] + list(pareto['pct_total'].cumsum())
x   = [0, 20, 40, 60, 80, 100]
axes[1].plot(x, cum, 'o-', color='#1a3a5c', lw=2.5, ms=8)
axes[1].axhline(80, color='red', ls='--', alpha=.6, label='80% Revenue Line')
axes[1].fill_between(x, cum, alpha=.12, color='#2e6da4')
axes[1].set_title('Cumulative Revenue Curve (Pareto)', fontsize=14, fontweight='bold', pad=15)
axes[1].set_xlabel('Cumulative % of Customers')
axes[1].set_ylabel('Cumulative % of Revenue')
axes[1].legend()

plt.tight_layout()
plt.savefig(os.path.join(r'project1-retail-analysis','pareto_analysis.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## Analysis 2 · Repeat Purchase Rate
**Business Question:** What % of customers bought more than once — and how much more do loyal buyers spend?

```
repeat_sql = '''
WITH freq AS (
    SELECT customer_id,
           COUNT(DISTINCT order_id)      AS num_orders,
           ROUND(SUM(order_total), 2)    AS ltv,
           ROUND(AVG(order_total), 2)    AS aov
    FROM   orders
    GROUP  BY customer_id
),
seg AS (
    SELECT *,
           CASE WHEN num_orders  = 1            THEN 'One-Time Buyer'
                WHEN num_orders BETWEEN 2 AND 4 THEN 'Occasional Buyer'
                ELSE                                 'Loyal Customer'
           END AS segment
    FROM freq
)
SELECT segment,
       COUNT(*)                                              AS customers,
       ROUND(COUNT(*)*100.0 / SUM(COUNT(*)) OVER (), 1)    AS pct_customers,
       ROUND(SUM(ltv), 2)                                   AS total_rev,
       ROUND(AVG(ltv), 2)                                   AS avg_ltv,
       ROUND(AVG(aov), 2)                                   AS avg_order_value
FROM   seg
GROUP  BY segment
ORDER  BY avg_ltv DESC
'''
repeat = pd.read_sql(repeat_sql, conn)
print(repeat.to_string(index=False))
loyal  = repeat[repeat['segment']=='Loyal Customer']['avg_ltv'].values
one_t  = repeat[repeat['segment']=='One-Time Buyer']['avg_ltv'].values
if loyal.size and one_t.size:
    print(f"\n>>> Key Finding: Loyal customers spend {loyal[0]/one_t[0]:.1f}x more on average than one-time buyers.")
```

```
ORDER_SEG = ['One-Time Buyer','Occasional Buyer','Loyal Customer']
rp = repeat.set_index('segment').reindex(ORDER_SEG).reset_index()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
colors = ['#89c4e1','#2e6da4','#1a3a5c']

axes[0].bar(rp['segment'], rp['customers'], color=colors)
axes[0].set_title('Customers per Purchase Segment', fontsize=14, fontweight='bold', pad=15)
axes[0].set_ylabel('Number of Customers')
for i, (_, row) in enumerate(rp.iterrows()):
    axes[0].text(i, row['customers']+3, f"{row['pct_customers']}%", ha='center', fontweight='bold')

axes[1].bar(rp['segment'], rp['avg_ltv'], color=colors)
axes[1].set_title('Average Lifetime Value per Segment', fontsize=14, fontweight='bold', pad=15)
axes[1].set_ylabel('Avg Lifetime Value ($)')
axes[1].yaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))
for i, (_, row) in enumerate(rp.iterrows()):
    axes[1].text(i, row['avg_ltv']+5, f"${row['avg_ltv']:,.0f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig(os.path.join(r'project1-retail-analysis','repeat_purchase.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## Analysis 3 · Monthly Revenue & Seasonality
**Business Question:** Which months and quarters drive the most revenue? Is there a Q4 holiday effect?

```
monthly_sql = '''
SELECT strftime('%Y',    order_date)    AS year,
       strftime('%m',    order_date)    AS month,
       strftime('%Y-%m', order_date)    AS ym,
       CASE WHEN CAST(strftime('%m', order_date) AS INT) BETWEEN 1 AND 3  THEN 'Q1'
            WHEN CAST(strftime('%m', order_date) AS INT) BETWEEN 4 AND 6  THEN 'Q2'
            WHEN CAST(strftime('%m', order_date) AS INT) BETWEEN 7 AND 9  THEN 'Q3'
            ELSE 'Q4' END              AS quarter,
       COUNT(DISTINCT order_id)        AS num_orders,
       ROUND(SUM(order_total),  2)     AS monthly_rev,
       ROUND(AVG(order_total),  2)     AS avg_order,
       ROUND(SUM(SUM(order_total)) OVER (
           PARTITION BY strftime('%Y', order_date)
           ORDER BY     strftime('%m', order_date)), 2) AS running_annual_rev
FROM   orders
GROUP  BY year, month
ORDER  BY ym
'''
monthly = pd.read_sql(monthly_sql, conn)
monthly['date'] = pd.to_datetime(monthly['ym'], format='%Y-%m')
print(monthly[['ym','num_orders','monthly_rev','quarter','running_annual_rev']].to_string(index=False))
```

```
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

for yr in sorted(monthly['year'].unique()):
    sub = monthly[monthly['year']==yr]
    axes[0].plot(sub['date'], sub['monthly_rev'], 'o-', lw=2.5, label=yr)
axes[0].set_title('Monthly Revenue Trend (2023–2024)', fontsize=14, fontweight='bold', pad=15)
axes[0].set_ylabel('Revenue ($)')
axes[0].yaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))
axes[0].legend()

quarterly = monthly.groupby(['year','quarter'])['monthly_rev'].sum().reset_index()
qp = quarterly.pivot(index='quarter', columns='year', values='monthly_rev')
qp.plot(kind='bar', ax=axes[1], color=['#1a3a5c','#89c4e1'], edgecolor='white', width=.7)
axes[1].set_title('Revenue by Quarter', fontsize=14, fontweight='bold', pad=15)
axes[1].set_ylabel('Revenue ($)')
axes[1].set_xlabel('Quarter')
axes[1].yaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig(os.path.join(r'project1-retail-analysis','monthly_trends.png'), dpi=150, bbox_inches='tight')
plt.show()
print(">>> Key Finding: Q4 (Oct–Dec) consistently drives peak revenue — holiday demand effect is clearly visible.")
```

## Analysis 4 · Product Category Performance
**Business Question:** Which categories lead in revenue, and which command the highest average transaction value?

```
cat_sql = '''
SELECT oi.category,
       COUNT(DISTINCT o.order_id)          AS orders,
       SUM(oi.quantity)                    AS units_sold,
       ROUND(SUM(oi.line_total), 2)        AS revenue,
       ROUND(AVG(oi.line_total), 2)        AS avg_line_val,
       ROUND(SUM(oi.line_total)*100.0 /
             SUM(SUM(oi.line_total)) OVER (), 1) AS rev_share_pct
FROM   order_items oi
JOIN   orders      o  ON oi.order_id = o.order_id
GROUP  BY oi.category
ORDER  BY revenue DESC
'''
cats = pd.read_sql(cat_sql, conn)
print(cats.to_string(index=False))
top = cats.iloc[0]
print(f"\n>>> Key Finding: {top['category']} leads with {top['rev_share_pct']}% of total revenue.")
```

```
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
col5 = ['#1a3a5c','#2e6da4','#4e9fd4','#89c4e1','#c6e2f0']

axes[0].barh(cats['category'], cats['revenue'], color=col5)
axes[0].set_title('Total Revenue by Category', fontsize=14, fontweight='bold', pad=15)
axes[0].set_xlabel('Revenue ($)')
axes[0].xaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))
for i, (_, row) in enumerate(cats.iterrows()):
    axes[0].text(row['revenue']+500, i, f"{row['rev_share_pct']}%", va='center', fontweight='bold')

axes[1].barh(cats['category'], cats['avg_line_val'], color=col5)
axes[1].set_title('Avg Line Item Value by Category', fontsize=14, fontweight='bold', pad=15)
axes[1].set_xlabel('Avg Value ($)')
axes[1].xaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))

plt.tight_layout()
plt.savefig(os.path.join(r'project1-retail-analysis','category_performance.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## Analysis 5 · Top 10 Customers by Lifetime Value (Window Functions)
**SQL concepts:** `RANK() OVER()`, `RANK() OVER(PARTITION BY ...)`, `MIN/MAX` aggregates

```
top10_sql = '''
WITH stats AS (
    SELECT o.customer_id,
           c.customer_city,
           COUNT(DISTINCT o.order_id)       AS total_orders,
           ROUND(SUM(o.order_total),  2)    AS lifetime_value,
           ROUND(AVG(o.order_total),  2)    AS avg_order,
           MIN(DATE(o.order_date))          AS first_order,
           MAX(DATE(o.order_date))          AS last_order,
           RANK() OVER (ORDER BY SUM(o.order_total) DESC)                    AS global_rank,
           RANK() OVER (PARTITION BY c.customer_city
                        ORDER BY SUM(o.order_total) DESC)                    AS city_rank
    FROM   orders    o
    JOIN   customers c ON o.customer_id = c.customer_id
    GROUP  BY o.customer_id, c.customer_city
)
SELECT * FROM stats ORDER BY global_rank LIMIT 10
'''
top10 = pd.read_sql(top10_sql, conn)
print(top10[['customer_id','customer_city','total_orders','lifetime_value',
             'avg_order','global_rank','city_rank']].to_string(index=False))
```

```
fig, ax = plt.subplots(figsize=(12, 6))
labels = [f"{r['customer_id']}\n({r['customer_city']})" for _, r in top10.iterrows()]
colors = [f'#{int(20+i*18):02x}{int(80+i*12):02x}{int(180-i*14):02x}' for i in range(10)]
bars   = ax.barh(labels, top10['lifetime_value'], color=colors)
ax.set_title('Top 10 Customers by Lifetime Value', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Lifetime Value ($)')
ax.xaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))
ax.invert_yaxis()
for b, v in zip(bars, top10['lifetime_value']):
    ax.text(b.get_width()+200, b.get_y()+b.get_height()/2, f'${v:,.0f}', va='center', fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(r'project1-retail-analysis','top10_customers.png'), dpi=150, bbox_inches='tight')
plt.show()
conn.close()
print("\n✓ Analysis complete. Database connection closed.")
```
